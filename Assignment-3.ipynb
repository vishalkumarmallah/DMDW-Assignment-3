```python
from matplotlib import pyplot
from pandas import read_csv
import numpy 
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age',
'class']
data=read_csv('student-mat.csv')
correlations = data.corr()
fig = pyplot.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,9,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
pyplot.show()
```

```python
import pandas as pd
csv_file = "student-mat.csv"
data = pd.read_csv(csv_file)
binary_attributes = ["failures", "absences"]  
numerical_attributes = ["age", "famsize"] 
nominal_attributes = ["guardian", "sex"] 
binary_matrix = data[binary_attributes].values
numerical_matrix = data[numerical_attributes].values
nominal_matrix = pd.get_dummies(data[nominal_attributes]).values
```
